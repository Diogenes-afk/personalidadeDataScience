```python
kmeans = KMeans(n_clusters=5)
k_fit=kmeans.fit(data)
```

### Atribuindo a quantidade de numero de clusteres a base inteira tratada e incluindo na variavel Kmeans

```python
predicoes = k_fit.labels_
data['Clusters'] = predicoes
```

```python
data.head()
```

```python
data["Clusters"].value_counts()
###visualizando a quantidade de respostas em cada grupo
```

```python
data.groupby('Clusters').mean()
```

```python
col_list = list(data)
ext_ = col_list[0:10]
est_ = col_list[10:20]
agr_ = col_list[20:30]
CON_ = col_list[30:40]
opn_ = col_list[40:50]
```

```python
data_soma = pd.DataFrame()
data_soma['Extrovertido'] = data[ext_].sum(axis=1)/10
data_soma['Neurotico'] = data[est_].sum(axis=1)/10
data_soma['Agradavel'] = data[agr_].sum(axis=1)/10
data_soma['Consciente'] = data[CON_].sum(axis=1)/10
data_soma['open'] = data[opn_].sum(axis=1)/10
data_soma['clusters'] = predicoes
```

```python
data_clusters = data_soma.groupby('clusters').mean()
```

```python
plt.figure(figsize=(22,3))
for i in range(0, 5):
    plt.subplot(1,5,i+1)
    plt.bar(data_clusters.columns, data_clusters.iloc[:, i], color='orange', alpha=1)
    plt.plot(data_clusters.columns, data_clusters.iloc[:, i], color='blue')
    plt.title('Grupo ' + str(i))
    plt.xticks(rotation=45)
    plt.ylim(0,4);
```

```python
import gradio as gr
```

```python
codebook = open("codebook.txt").read().split("\n")
```

```python
codebook
```

```python
questions_=[]
for q in codebook:
    q = str(q)
    questions_.append(q[q.find("\t"):].lstrip())
```

```python
questions_
```

```python
input_questions = []
for q in questions_:
    obj_input = gr.inputs.Slider(minimum=1, maximum=5, step=1, default=3, label=q)
    input_questions.append(obj_input)
```

```python
input_questions
```

```python
import gradio as gr
def predict(*outputs_questions):
    outputs_questions = np.array(outputs_questions).reshape(1, -1)
    return k_fit.predict(outputs_questions)

iface = gr.Interface(
    fn = predict,
    title = "Teste personalidade Didico.com",
    description = "Sistema de identifica√ß√£o de personalidade para RH",
    inputs = input_questions,
    outputs = "text")

iface.launch(share=True)
```
